```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generate Data
X= np.arange(10)
print(X)
Y=(X-5)**2 +3
print(Y)


```

```python
#Visualize
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#gradient descent
# x = x - ηx' , η - > learning rate ;
#init x with any random value
x =0
lr = 0.1

plt.plot(X,Y)
for i in range(50):
    grad = 2*(x-5)
    x = x - lr *grad
    y = (x-5)**2 + 3
    plt.scatter(x,y)
    print(x)

plt.show()
```
